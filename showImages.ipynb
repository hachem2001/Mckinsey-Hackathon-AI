```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os
```

```python
# Open the image
def visualize_tif_images(folder_path, n):
    # Iterate through all files in the folder
    count = 0
    for filename in os.listdir(folder_path):
        if ("id_6675" == filename[-11:-4]) & ("20230223" == filename[0:8]) :#Choose a specific file
            if(count >= n):
                break;
            count += 1;
            img_path = os.path.join(folder_path, filename)
            img = mpimg.imread(img_path)

            # Display the image
            plt.imshow(img)
            plt.show()
```

```python
def visualize_tif_images2(folder_path, n):
    # Iterate through all files in the folder
    count = 0
    for filename in os.listdir(folder_path):
        if ("id_6675" == filename[-17:-10]) & ("20230223" == filename[0:8]) : #Choose a specific file
            if(count >= n):
                break;
            count += 1;
            img_path = os.path.join(folder_path, filename)
            img = mpimg.imread(img_path)

            # Display the image
            plt.imshow(img)
            plt.show()
```

```python
print("Original Image:")
visualize_tif_images("./images/plume", 1)
```
